```python
from matplotlib import animation as v_anim
import matplotlib.pyplot as plt
import random as rd
import numpy as np
from IPython.display import HTML
import time as take_time
```

```python
class Short_Anim():
    def __init__(self, ls_input, linsp_a_z, linsp_len, scope_lim, f_var, i_var):
        (self.ls_input, self.linsp_len, self.f_var, self.i_var) = ls_input, linsp_len, f_var, i_var
        self.coords = plt.axes(projection='3d')
        self.coords.axis('off')
        self.linsp_len = linsp_len
        self.coords.set_xlim((-scope_lim,scope_lim))
        self.coords.set_ylim((-scope_lim,scope_lim))
        self.coords.set_zlim((-scope_lim,scope_lim))
        self.t_var = np.linspace(linsp_a_z[0], linsp_a_z[1], linsp_len)
        self.anim = v_anim.FuncAnimation(self.coords.figure, self.animate, frames=self.f_var, interval=self.i_var)
    def animate(self, i_var):
        self.coords = plt.axes(projection='3d')
        
        x_var = np.outer(self.t_var, np.ones(self.linsp_len)) #/  (i_var+1)
        y_var = np.outer(self.t_var, np.ones(self.linsp_len)).T #/ (i_var+1)
        z_var =  -abs(x_var**7+ y_var**11 ) ** (1/((i_var/100)+.01)) #*   (i_var/100)
        self.coords.plot_surface(x_var, y_var, z_var, cmap='Spectral', edgecolor='none', alpha = 0.7)
        self.coords.axis('off')
        self.coords.view_init(elev=25, azim=-45+(1+i_var*2)) # +(1+i_var*2)
    def g_render(self):
        return HTML(self.anim.to_jshtml())    
    def save_anim(self, f_name='0'):
        self.anim.save('{}.gif'.format(f_name), dpi=330)
```

```python
time_start = take_time.time()
h_c_v =['midnightblue', 'rebeccapurple','slateblue', 'navy', 'indigo', 'black', 'darkslateblue']
short_a = Short_Anim([11, 13],  # 'ls_input'
                     (-1,1),            # 'linsp_a_z'
                     25,                           # 'linsp_len'
                     1,                           # 'scope_lim'
                     150,                            # 'f_var'
                     75)                            # 'i_var'
#short_a.g_render()
short_a.save_anim('vulcan2')
```

```python
time_stop = take_time.time()
time_spent = time_stop - time_start
print(time_spent)
```
